```python
import pandas as pd
import numpy as np
```

```python
austin_df = pd.read_csv("Resources/austin_2018_raw.csv")
austin_df.head()
```

```python
austin_df["BUDGET"]= (austin_df['BUDGET'].replace( '[\$,)]','', regex=True ).replace( '[(]','-',   regex=True ).astype(float))
austin_df["EXPENDITURES"]= (austin_df['EXPENDITURES'].replace( '[\$,)]','', regex=True ).replace( '[(]','-',   regex=True ).astype(float))
```

```python
austin_df2 = austin_df.loc[austin_df["BUDGET_FISCAL_YEAR"]==2018]
austin_df2.head()
```

```python
austin_df3 = austin_df2.groupby(["DEPT_ROLLUP_NAME"]).sum()
austin_df3.head()
```

```python
austin_df4 = austin_df3[["EXPENDITURES"]]
austin_df4
```

```python
classification = [6, 4, 2, 2,7, 1, 2, 5, 7, 7, 7, 4, 3, 3, 3, 3, 1, 3, 3, 6, 6, 3, 5, 4, 4, 4, 4, 4, 4, 5, 8, 5, 4, 4, 1, 2, 5, 6, 7, 2, 3, 7]
```

```python
department_name = []

for x in classification:
    if x == 1:
        department_name.append("Community Health")
    if x == 2:
        department_name.append("Culture & Recreation")
    if x == 3:
        department_name.append("General  Administration & Finance")
    if x==4:
        department_name.append("General City Responsibilities")
    if x ==5:
        department_name.append("Human Welfare & Neighborhood Development")
    if x==6:
        department_name.append("Public Protection")
    if x ==7:
        department_name.append("Public Works/Transportation & Commerce")
    if x ==8:
        department_name.append("Other")
len(department_name)
```

```python
austin_df4["Department"] = department_name
```

```python
austin_df4
```

```python
austin_df5 = austin_df4.groupby(["Department"]).sum()
austin_df5.rename(columns = {"EXPENDITURES":"Amount"})
```

```python
austin_df5.to_csv("Transformed/Austin.csv")
```
