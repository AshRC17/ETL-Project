```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sanfran_df = pd.read_csv("Resources/sanfrancisco_raw.csv")
sanfran_df.head()
```

```python
sanfran_df["Amount"] = sanfran_df["Vouchers Paid"]+ sanfran_df["Vouchers Pending"]
```

```python
sanfran_df2 = sanfran_df.loc[sanfran_df["Fiscal Year"]==2018]

sanfran_df2.head()
```

```python
sanfran_df3 = sanfran_df2.groupby(["Organization Group"]).sum()


sanfran_df3
```

```python
sanfran_df4 = sanfran_df3[["Amount"]]

sanfran_df4
```

```python
sanfran_df5 = sanfran_df4.rename_axis("Department")
sanfran_df5
```

```python
sanfran_df5.to_csv("Transformed/SanFrancisco.csv")
```
