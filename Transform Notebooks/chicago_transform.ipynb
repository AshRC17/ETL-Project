```python
import pandas as pd
import numpy as np
```

```python
chicago_df = pd.read_csv("Resources/chicago_2018_raw.csv")
chicago_df.head()
```

```python
chicago_df2 = chicago_df.groupby(["DEPARTMENT DESCRIPTION"]).sum()
chicago_df2
```

```python
chicago_df3 = chicago_df2[["2018 ORDINANCE (AMOUNT $)"]]
chicago_df3
```

```python
print(len(chicago_df3))
classification = [3, 6, 7, 3, 4, 7, 6, 3, 4, 6, 6, 2, 2, 5, 3, 3, 3, 7, 7, 4, 4, 3, 3, 3, 1, 3, 3, 4, 4, 5,  3, 6, 3, 5, 6, 3]
print(len(classification))
```

```python
department_name = []

for x in classification:
    if x == 1:
        department_name.append("Community Health")
    if x == 2:
        department_name.append("Culture & Recreation")
    if x == 3:
        department_name.append("General  Administration & Finance")
    if x==4:
        department_name.append("General City Responsibilities")
    if x ==5:
        department_name.append("Human Welfare & Neighborhood Development")
    if x==6:
        department_name.append("Public Protection")
    if x ==7:
        department_name.append("Public Works/Transportation & Commerce")
    if x ==8:
        department_name.append("Other")
```

```python
chicago_df3["Department"] = department_name
chicago_df3
```

```python
chicago_df4 = chicago_df3.groupby(["Department"]).sum()
chicago_df4
```

```python
chicago_df5 = chicago_df4.rename(columns = {"2018 ORDINANCE (AMOUNT $)":"Amount"})
chicago_df5
```

```python
chicago_df5.to_csv("Transformed/Chicago.csv")
```
