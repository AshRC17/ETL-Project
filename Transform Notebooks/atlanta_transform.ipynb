```python
import pandas as pd
import numpy as np
```

```python
atlanta_df = pd.read_csv("Resources/atlanta_raw.csv")
atlanta_df.head()
```

```python
atlanta_df2 = atlanta_df.loc[atlanta_df["Fiscal Year"]==2018]
atlanta_df2.head()
```

```python
atlanta_df2["Department"].unique()
```

```python
atlanta_df3 = atlanta_df2.groupby(["Department"]).sum().rename_axis("Departments")

atlanta_df3.head()
```

```python
atlanta_df4 = atlanta_df3[[ "Amount"]]
atlanta_df4
```

```python
print(len(atlanta_df4))
classification = [4, 4, 4, 7, 6, 4, 3, 6, 3, 3, 4, 6, 4, 4,  7, 4, 7, 2, 5, 4, 4, 8]
len(classification)
```

```python
department_name = []

for x in classification:
    if x == 1:
        department_name.append("Community Health")
    if x == 2:
        department_name.append("Culture & Recreation")
    if x == 3:
        department_name.append("General  Administration & Finance")
    if x==4:
        department_name.append("General City Responsibilities")
    if x ==5:
        department_name.append("Human Welfare & Neighborhood Development")
    if x==6:
        department_name.append("Public Protection")
    if x ==7:
        department_name.append("Public Works/Transportation & Commerce")
    if x ==8:
        department_name.append("Other")
```

```python
atlanta_df4["Department"] = department_name
```

```python
atlanta_df5 = atlanta_df4.groupby(["Department"]).sum()
```

```python
atlanta_df5
```

```python
atlanta_df5.to_csv("Transformed/Atlanta.csv")
```
