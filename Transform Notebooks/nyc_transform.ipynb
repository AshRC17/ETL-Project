```python
import pandas as pd
import numpy as np
```

```python

nyc_df = pd.read_csv("Resources/nyc_2018_raw.csv")


nyc_df.head()
```

```python
nyc_df2 = nyc_df.groupby(["Agency Name"]).sum()
```

```python
nyc_df2["Budget"] = nyc_df2["Modified_Budget_ FY18 ($000,000)"]*(1000000)
nyc_df3 = nyc_df2[["Budget"]]
nyc_df3
```

```python
print(len(nyc_df3))
classification = [5, 4, 2, 3, 5, 4, 1, 5, 4, 4, 3, 3, 6, 2, 4, 5, 6, 6, 3, 1, 5, 3, 6, 2, 6, 3, 7, 4, 7, 5, 5, 6, 1, 5, 4, 2, 4, 2, 2, 4, 6, 2, 1]
len(classification)
```

```python
department_name = []

for x in classification:
    if x == 1:
        department_name.append("Community Health")
    if x == 2:
        department_name.append("Culture & Recreation")
    if x == 3:
        department_name.append("General  Administration & Finance")
    if x==4:
        department_name.append("General City Responsibilities")
    if x ==5:
        department_name.append("Human Welfare & Neighborhood Development")
    if x==6:
        department_name.append("Public Protection")
    if x ==7:
        department_name.append("Public Works/Transportation & Commerce")
    if x ==8:
        department_name.append("Other")
```

```python
nyc_df3["Department"] = department_name
nyc_df3
```

```python
nyc_df4 = nyc_df3.groupby(["Department"]).sum()
nyc_df5 = nyc_df4.rename(columns = {"Budget":"Amount"})
nyc_df5
```

```python
nyc_df5.to_csv("Transformed/NYC.csv")
```
